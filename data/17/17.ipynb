```python
import pandas as pd

df = pd.read_csv('raw.csv', sep=',', header=None, names=['uni', 'hos', 'dptm', 'name', 'img_name', 'position', 'clinic', 'edu', 'career', 'study', 'etc'])
df = df.drop(0).reset_index(drop=True)
df
```

```python
# v.2

edu_text = ''
etc_text = ''

for i in range(len(df)):
    text = str(df.at[i, 'career'])
    rows = text.split('\r\n')

    for row in rows:
        if row.startswith('[학력]'):
            edu_text += row + '\r\n'
        elif row.startswith('[수상]'):
            etc_text += row + '\r\n'

    if edu_text:
        df.at[i, 'edu'] = edu_text.strip()
        edu_text = ''
    if etc_text:
        df.at[i, 'etc'] = etc_text.strip()
        etc_text = ''

    # Remove the rows from 'career' that were moved to 'edu' and 'etc'
    df.at[i, 'career'] = '\r\n'.join([row for row in rows if not row.startswith('[학력]') and not row.startswith('[수상]')])

df.to_csv('output.csv', sep='\t', index=False)
df
```

```python
# v.2.1 경력 앞 줄바꿈 삭제

for i in range(len(df)):
    text = str(df.at[i, 'career'])

    # Remove the first '\r\n' if it exists
    if text.startswith('\r\n'):
        text = text[2:]

    df.at[i, 'career'] = text

df.to_csv('output.csv', sep='\t', index=False)
df
```
