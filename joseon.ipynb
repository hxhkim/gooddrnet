```python
departments = [
    '감염내과',
    '내분비대사내과',
    '류마티스내과',
    '소화기내과',
    '순환기내과',
    '신장내과',
    '종양혈액내과',
    '호흡기내과',
    '가정의학과',
    '마취통증의학과',
    '방사선종양학과',
    '병리과',
    '비뇨의학과',
    '산부인과',
    '성형외과',
    '소아청소년과',
    '신경과',
    '신경외과',
    '심장혈관흉부외과',
    '안과',
    '영상의학과',
    '외과',
    '응급의학과',
    '이비인후과',
    '재활의학과',
    '정신건강의학과',
    '정형외과',
    '직업환경의학과',
    '진단검사의학과',
    '피부과',
    '핵의학과'
]
```

```python
# 완성 ver.

all_doctor_info_list = []
doctor_info_list = []

for department in departments:
    doctor_info_list = []

    url_dptm = f"https://hosp.chosun.ac.kr/medi_depart/?site=hospital&mn=471&type=doctor&catename={department}"

    r_dptm = requests.get(url=url_dptm)
    html_dptm = r_dptm.content
    soup_dptm = BeautifulSoup(html_dptm, 'html.parser')
    hover = soup_dptm.find_all('a', class_="btn_more")

    for i in range(len(hover)):

        hover = soup_dptm.find_all('a', class_="btn_more")[i]
        link = hover['href']
        url_doc = f"https://hosp.chosun.ac.kr{link}"
        
        r_doc = requests.get(url=url_doc)
        html_doc = r_doc.content
        soup = BeautifulSoup(html_doc, 'html.parser')

        img_src = soup.find('div', class_="profile_img").find('img')['src'] if soup.find('div', class_="profile_img") else 'none'
        img_name = img_src.replace('/data/doctor/', '') if 'noimage' not in img_src else ''
        img_url = f"https://hosp.chosun.ac.kr{img_src}"

        tag2 = soup.find('div', class_="profile_txt")

        if tag2:
            dptm = tag2.find('p', class_="pro_dep").text
            name, position = tag2.find('p', class_="pro_name").text.split(' ')
            clinic = tag2.find('p', class_="pro_subject").text
        
        else:
            dptm = ""
            name = ""
            position = ""
            clinic = ""
        
        tag3 = soup.find('div', class_="pro_career")
        
        if tag3:
            career_items = tag3.find('ul').find_all('li')
            career = "\n".join(item.get_text(strip=True) for item in career_items)
        else:
            career = ""

        try:
            response = requests.get(img_url)
            if response.status_code == 200:
                image_data = response.content
                image = Image.open(BytesIO(image_data))
                image.save(f"joseon/{img_name}")
            else:
                print(f"이미지 다운로드 실패. HTTP 상태 코드: {response.status_code}")
        except Exception as e:
            print(f"이미지 다운로드 중 오류 발생: {str(e)}")
            continue      
        
        doctor_info = {
            'dptm': dptm,
            'name': name,
            'img_name': img_name, 
            'position': position,
            'clinic': clinic,
            'career': career
        }
        doctor_info_list.append(doctor_info)
    all_doctor_info_list.extend(doctor_info_list)

all_doctor_info_list

with open("joseon/doctor_info.csv", 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['dptm', 'name', 'img_name', 'position', 'clinic', 'career']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for doctor_info in all_doctor_info_list:
        writer.writerow(doctor_info)
```
