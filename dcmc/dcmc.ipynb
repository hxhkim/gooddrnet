```python
dptms_dict = {}

for j in range(len(medical_list)):
    dptm_names = medical_list[j].find('p', class_='tit').text
    dptm_links = medical_list[j].find('a', string='의료진소개')['href']

    dptms_info = {
        dptm_names : dptm_links
    }
    dptms_dict.update(dptms_info)
dptms_dict
```

```python
for department in departments_to_remove:
    if department in dptms_dict:
        del dptms_dict[department]
```

### 의료진 정보 모으기

```python
# 완성 ver.

all_doctor_info_list = []
doctor_info_list = []

for key in dptms_dict:
    doctor_info_list = []

    url_dptm = f"https://www.dcmc.co.kr{dptms_dict[key]}"
    r_dptm = requests.get(url=url_dptm)
    html_dptm = r_dptm.content
    soup_dptm = BeautifulSoup(html_dptm, 'html.parser')

    box = soup_dptm.find_all('div', class_='box')

    for j in range(len(box)):
        link = box[j].find('a', class_='rev_btn white')['href']
        url_doc = f"https://www.dcmc.co.kr{link}"

        r_doc = requests.get(url=url_doc)
        html_doc = r_doc.content
        soup = BeautifulSoup(html_doc, 'html.parser')

        img_src = box[j].find('p', class_='photo').find('img')['src'] if box[j].find('p', class_='photo') else 'none'
        img_name = img_src.replace('http://www.dcmc.co.kr/data/dr_list/', '') if 'no' not in img_src else 'none'
        
        tag1 = soup.find('div', class_="d_career")
        if tag1:
            title_tag = tag1.find_all('h3', class_='tit h3_t')

            tmp_dict = {}

            for i in range(len(title_tag)):

                title_tag = tag1.find_all('h3', class_='tit h3_t')
                title = title_tag[i].text
                contents_tag = tag1.find_all('ul', class_='list')[i].find_all('li')
                contents = "\n".join(item.get_text(strip=True) for item in contents_tag)

                tmp_info = {
                    title: contents
                }
                tmp_dict.update(tmp_info)
        else:
            tmp_dict = {}

        tag2 = soup.find('div', class_="profile")
        if tag2:
            dptm = tag2.find('p', class_='team').text
            name = tag2.find('p', class_='name').text
            position_tmp = tag2.find('dl', class_='position').find('dd').text
            position = position_tmp.split('/')[0] if '/' in position_tmp else position_tmp
            clinic = tag2.find('dl', class_='part').find('dd').text
        else:
            dptm = ""
            name = ""
            position = ""
            clinic = ""

        try:
            response = requests.get(img_src)
            if response.status_code == 200:
                image_data = response.content
                image = Image.open(BytesIO(image_data))
                image.save(f"dcmc/img/{img_name}")
            else:
                print(f"{img_name}이미지 다운로드 실패. HTTP 상태 코드: {response.status_code}")
        except Exception as e:
            print(f"이미지 다운로드 중 오류 발생: {doctor_info['name']} ({doctor_info['dptm']}) - {str(e)}")
            continue  

        doctor_dict = {
            'dptm': dptm,
            'name': name,
            'img_name' : img_name,
            'position': position,
            'clinic': clinic,
        }
        doctor_dict

        doctor_info = {**doctor_dict, **tmp_dict}
        doctor_info

        doctor_info_list.append(doctor_info)
    all_doctor_info_list.extend(doctor_info_list)

all_doctor_info_list

with open("dcmc/doctor_info.csv", 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['dptm', 'name', 'img_name', 'position', 'clinic', '학력사항', '경력사항', '학회활동', '수상이력', '해외연수']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for doctor_info in all_doctor_info_list:
        # 딕셔너리에 각 필드가 존재하는지 확인하고, 없다면 빈 문자열로 설정합니다.
        for field in fieldnames:
            if field not in doctor_info:
                doctor_info[field] = ''

        # 필드명이 딕셔너리에 있는지 확인하고 없다면 추가합니다.
        for field in doctor_info:
            if field not in fieldnames:
                fieldnames.append(field)
                writer.fieldnames = fieldnames
        
        writer.writerow(doctor_info)
```
